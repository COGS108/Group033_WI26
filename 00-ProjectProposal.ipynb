## Research Question

Using the ExtraSensory minute-level in-the-wild dataset, is social context (alone vs with_friends vs with_family) associated with the probability that a person is physically active in that same minute? We will define the outcome active_minute = 1 if any of the labels walking, running, or bicycling is present for that minute, and 0 otherwise. The main predictors will be the social labels (alone, with_friends, with_family), and we will adjust for confounds using time-of-day, day-of-week, holiday indicator, local weather (temperature and precipitation), and environment context (for example at_home, at_work, outside). We will estimate associations using a mixed-effects logistic regression with a participant random intercept to account for repeated measures within individuals. Primary metrics will be odds ratios and marginal probability differences between social contexts with ninety-five percent confidence intervals, plus sensitivity analyses using alternative definitions of active_minute.


## Background and Prior Work

Physical activity is a core public-health behavior, but it is not only an individual choice—people’s activity levels are shaped by *context*, including **who they are with**. Social and behavioral research has repeatedly argued that social connection and support can influence whether people start and maintain physical activity, but the strength and direction of this relationship can vary depending on the *source* of the social influence (friends versus family) and the situation.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) This motivates our focus on separating `with_friends` and `with_family` rather than treating “being social” as one generic category.

Prior work also suggests that **friends and family are not interchangeable** in how they relate to physical activity. For example, Larsen et al. explicitly examined differences between **family integration** and **friend integration** and their association with physical activity, treating them as distinct social dimensions rather than one combined predictor.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) Similarly, Sarkar et al. compared different sources of social support (family, friends, coworkers) and analyzed how these relate to moderate-to-vigorous physical activity, reinforcing the idea that the “type” of social context matters for activity outcomes.<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3) These studies support our project design choice: we will estimate separate associations for `with_friends` and `with_family` relative to `alone`.

However, much of the behavioral literature relies on surveys, retrospective recall, or diaries, which can miss the fine-grained, moment-to-moment nature of daily behavior. Time-diary approaches like the American Time Use Survey (ATUS) are valuable because they record what people did and also include “with whom” context, enabling large-scale analyses of social context and activity patterns across the population.<a name="cite_ref-4"></a>[<sup>4</sup>](#cite_note-4) For example, Sener et al. used ATUS to examine the timing, location, duration, and **social context** of physically active behavior, demonstrating that the “with whom” dimension is a meaningful lens for understanding activity in real life.<a name="cite_ref-5"></a>[<sup>5</sup>](#cite_note-5) These diary-based results motivate our project’s external-validity strategy: use an in-the-wild sensing dataset for minute-level inference, and use diary evidence as a broader contextual check.

On the data science side, mobile and wearable sensing offers a way to study behavior in natural settings—but it is difficult because real-world behavior is unscripted and phone usage is unconstrained. The ExtraSensory project directly addresses this by providing a large-scale **in-the-wild** dataset with minute-level, multi-label context annotations (including social context labels) paired with smartphone/smartwatch sensor data, specifically designed to study detailed human context outside the lab.<a name="cite_ref-6"></a>[<sup>6</sup>](#cite_note-6) Because ExtraSensory includes both social labels and activity labels at the same time resolution, it enables our proposed **statistical inference** approach: estimating the association between being with friends or family and the likelihood of being active, while controlling for time structure and environment. Our work therefore connects behavior-science theory (social context matters) with a measurement approach suited to real life (minute-level in-the-wild context labels), and aims to quantify whether the association holds after adjustment for major confounds.

## Hypothesis


We hypothesize that minutes labeled with_friends will have a higher probability of active_minute = 1 than minutes labeled alone, even after controlling for time-of-day, day-of-week, environment, weather, and holidays. We also predict the association for with_family will be smaller and less consistent (closer to zero on average), because “with family” includes both active contexts (errands/outings) and sedentary contexts (meals and time at home).

## Data


1) Ideal dataset (what we would want to answer the question best)

The ideal dataset would be minute-level repeated measures for a large sample of people, where each row represents (participant_id, timestamp_minute) and includes: (1) social context with clear categories (alone / with friends / with family), (2) a validated physical activity measure per minute (for example step count or energy expenditure from a wearable), (3) environment context (home/work/outdoors/transit), and (4) time variables (time-of-day, day-of-week, date). It would also include major confounders like weather (temperature/precipitation) and a holiday indicator, plus participant-level demographics and device-compliance flags. To estimate within-person associations reliably, we would aim for at least 100+ participants with 2–4 weeks of data each (millions of minute-observations total), collected in-the-wild using smartphones/wearables plus low-burden ecological momentary assessment for “who are you with.” Data would be stored as a tidy minute-level table plus a participant metadata table, with explicit missingness indicators for any absent labels or sensors.


2) Potential real datasets (multiple sources, each supporting a specific project need)

A) ExtraSensory (primary dataset: social context + activity + environment)
Where located / access: Official dataset website: https://extrasensory.ucsd.edu/ (downloadable for research use; follow the dataset’s citation/usage instructions).
Important variables we would use: minute-level labels for social context (alone, with_friends, with_family), minute-level activity labels (walking, running, bicycling) to construct active_minute, environment labels (for example at_home, at_work, outside), timestamps, and participant identifiers.
How it supports the project: This dataset directly contains both the predictor (social context) and outcome (activity) at the same time resolution, plus the key controls needed for our mixed-effects inference model.

B) American Time Use Survey (ATUS) (secondary dataset: external validity / triangulation)
Where located / access: U.S. Bureau of Labor Statistics ATUS portal + documentation: https://www.bls.gov/tus/ (publicly available microdata and user guides).
Important variables we would use: time-diary “episodes” with activity codes (to identify physical activity/exercise), time-of-day/date, and “with whom” context fields (to operationalize alone vs with others; where possible, family vs non-family).
How it supports the project: ATUS provides a large, population-based reference to check whether social context and physical activity co-occur in a different measurement framework (diary episodes rather than sensors), strengthening generalizability of our findings.

C) NOAA Integrated Surface Database (ISD / Global Hourly) (supporting dataset: weather controls)
Where located / access: NOAA NCEI ISD documentation and downloads: https://www.ncei.noaa.gov/products/land-based-station/integrated-surface-database (public access).
Important variables we would use: hourly temperature and precipitation (and possibly wind), matched by timestamp and nearest station / region.
How it supports the project: Weather influences both activity and social plans, so including weather reduces confounding and helps isolate the association between social context and activity.

D) Public holiday calendar (supporting dataset: day-type controls)
Where located / access: Microsoft Azure Open Datasets “Public Holidays”: https://learn.microsoft.com/en-us/azure/open-datasets/dataset-public-holidays (public dataset documentation).
Important variables we would use: date-level holiday indicators (holiday vs non-holiday; region where applicable).
How it supports the project: Holidays can systematically change both social context (more family/friends time) and activity routines, so controlling for holidays improves the validity and interpretability of estimated social-context effects.
